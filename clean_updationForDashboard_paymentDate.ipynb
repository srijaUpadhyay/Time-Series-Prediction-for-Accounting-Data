```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('TIH_cleaned.csv', parse_dates=['INVOICE_DT', 'Payment_Date'])
df.head()
```

```python
df['Payment_Date'] = pd.to_datetime(df['Payment_Date'], format='%d-%m-%Y', errors='coerce')
df['INVOICE_DT'] = pd.to_datetime(df['INVOICE_DT'], format='%d-%m-%Y', errors='coerce')
```

```python
df.dtypes
```

```python
df['Invoice_Month_Year'] = df['INVOICE_DT'].dt.to_period('M')
df['Payment_Month_Year'] = df['Payment_Date'].dt.to_period('M')
```

```python
df.tail()
```

```python
df_exp_month = df.groupby(['EXPENSE HEAD', 'Payment_Month_Year'])['GROSS AMOUNT'].sum()
```

```python
df_exp_month = df_exp_month.reset_index()
df_exp_month.head()
```

```python
df.to_csv('clean2_temp.csv', index=False)
```

```python
df_exp_month.to_csv('Expense_Monthly_PaymentDate.csv')
```

### Weekly Aggregation

```python
df['Payment_Week'] = df['Payment_Date'].dt.to_period('W')
df.head()
```

```python
df_agg_weekly = df.groupby(['EXPENSE HEAD', 'Payment_Week'])['GROSS AMOUNT'].sum()
df_agg_weekly = df_agg_weekly.reset_index()
df_agg_weekly
```

```python
df_agg_weekly.to_csv('WeeklyGrossAmtByExpenseHead.csv')
```
