```python
df.iloc[365 ,[6,7,8]] = [0,0,0] #6,7,8 denote columns CGST, SGST, IGST respectively
df.isnull().sum()
```

```python
df.loc[(df['P TAX (-)'].isnull() == True) & (df['TDS AMOUT (-)'].isnull() == True) & (df['GROSS AMOUNT'] - df['NET AMOUNT PAID '] >1),
    ['INVOICE_NO', 'GROSS AMOUNT', 'P TAX (-)', 'TDS AMOUT (-)', 'NET AMOUNT PAID ']]
```

```python
df.loc[(df['TDS AMOUT (-)'].isnull() == True) & (df['TDS PERCENTAGE'].isnull() != True), 
    ['INVOICE_NO', 'GROSS AMOUNT', 'P TAX (-)', 'TDS AMOUT (-)', 'NET AMOUNT PAID ', 'TDS PERCENTAGE']]
#No row where percentage is mentioned but TDS amount is NaN.
```

```python
df.loc[(df['TDS AMOUT (-)'].isnull() != True) & (df['TDS PERCENTAGE'].isnull() == True), 
    ['INVOICE_NO', 'GROSS AMOUNT', 'P TAX (-)', 'TDS AMOUT (-)', 'NET AMOUNT PAID ', 'TDS PERCENTAGE']]
#There are many rows with no TDS Percentage mentioned, but TDS Amount is deducted. 
#So imputing TDS Percent as 0 will not be correct in cases where TDS Amount is not Zero.
```

```python
df.loc[(df['P TAX (-)'].isnull() == True), 'P TAX (-)'] = 0
df.loc[(df['TDS AMOUT (-)'].isnull() == True), 'TDS AMOUT (-)'] = 0
df.loc[(df['TDS AMOUT (-)'] == 0), 'TDS PERCENTAGE'] = 0 #Imputing 0 only where TDS Amount is 0.
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.loc[(df['TDS PERCENTAGE'] == 'AS PER SLAB'), 
    ['INVOICE_NO', 'INVOICE_VALUE_WITHOUT_GST_OR_TDS', 'GROSS AMOUNT', 'P TAX (-)', 'TDS AMOUT (-)', 'NET AMOUNT PAID ', 'TDS PERCENTAGE']].head(7)
```

```python
df.loc[df['INVOICE_VALUE_WITHOUT_GST_OR_TDS']==0,:]
```

```python
#Calculate the TDS Percentage where the value is 'AS PER SLAB' or Null
df.loc[(df['TDS PERCENTAGE'] == 'AS PER SLAB') & (df['INVOICE_VALUE_WITHOUT_GST_OR_TDS']!=0), 'TDS PERCENTAGE'] =  \
    (df.loc[(df['TDS PERCENTAGE'] == 'AS PER SLAB') & (df['INVOICE_VALUE_WITHOUT_GST_OR_TDS']!=0), 'TDS AMOUT (-)'] \
     / df.loc[(df['TDS PERCENTAGE'] == 'AS PER SLAB') & (df['INVOICE_VALUE_WITHOUT_GST_OR_TDS']!=0), \
     'INVOICE_VALUE_WITHOUT_GST_OR_TDS']) *100
```

```python
df.loc[(df['TDS PERCENTAGE'].isnull() == True) & (df['INVOICE_VALUE_WITHOUT_GST_OR_TDS']!=0), 'TDS PERCENTAGE'] =  \
    (df.loc[(df['TDS PERCENTAGE'].isnull() == True) & (df['INVOICE_VALUE_WITHOUT_GST_OR_TDS']!=0), 'TDS AMOUT (-)'] \
     / df.loc[(df['TDS PERCENTAGE'].isnull() == True) & (df['INVOICE_VALUE_WITHOUT_GST_OR_TDS']!=0), \
     'INVOICE_VALUE_WITHOUT_GST_OR_TDS']) *100
```
